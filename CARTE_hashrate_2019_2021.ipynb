```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from plotly.graph_objs import Scatter, Figure, Layout
import plotly
import json
import folium
```

```python
#on charge les contours des pays 
countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

path_pays="https://raw.githubusercontent.com/GeoffroyENSAE/Cryptomonnaies-et-ecologie/main/hashrate_country_2019_2021.csv"

#on charge les donn√©es sur le Hashrate par pays
df=pd.read_csv(path_pays,sep=",")
df=df.replace('Mainland China','China')
df=df.replace('Russian Federation','Russia')
df=df.replace('Iran, Islamic Rep.','Iran')
df=df.replace('Ireland *','Ireland')
df=df.replace('Germany *','Germany')
```

```python
df['monthly_absolute_hashrate_EH/S']=df['monthly_absolute_hashrate_EH/S']*(10^18)
df_carte_2019=df[df['date']=="2019-12-01"].reset_index(drop=True)
df_carte_2021=df[df['date']=="2021-08-01"].reset_index(drop=True)
```

```python
# on cr√©e une carte vierge 
m = folium.Map(location=[40, -95], zoom_start=2)
# on gen√®re la carte √† deux couches  
folium.Choropleth(
  geo_data=countries.to_json(),
    data=df_carte_2019,
    columns=['country','monthly_absolute_hashrate_EH/S'],
    key_on="properties.name", 
     fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=1,
    legend_name='Hashrate mensuel en H/s',name='Hashrate D√©cembre 2019',
    smooth_factor=0).add_to(m)


folium.Choropleth(
  geo_data=countries.to_json(),
    data=df_carte_2021,
    columns=['country','monthly_absolute_hashrate_EH/S'],
    key_on="properties.name", 
     fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=1,
    legend_name='Hashrate mensuel en H/s',name='Hashrate Ao√ªt 2021',
    smooth_factor=0).add_to(m)
folium.LayerControl().add_to(m)
m
```

En cliquant sur le sigle avec des couches, on peut afficher la carte des Hashrates en 2019 ou bien en 2021. Par d√©faut les deux couches se superposent mais il faut n'en garder qu'une pour lire bien la carte.
